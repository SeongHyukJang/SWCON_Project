```python
import os
import json
from matplotlib import pyplot as plt
```

```python
os.chdir('./client')
print(os.getcwd())
```

```python
with open('LoadTestResult.json') as file:
    data = json.load(file)
    
py_data = data['OS']['Linux']['Server']['python']
js_data = data['OS']['Linux']['Server']['javascript']

####################### JSON GET #############################
py_json_get = py_data['resource']['json']['method']['GET'][0]
js_json_get = js_data['resource']['json']['method']['GET'][0]
##############################################################
####################### JSON POST ############################
py_json_post = py_data['resource']['json']['method']['POST'][0]
js_json_post = js_data['resource']['json']['method']['POST'][0]
##############################################################
####################### Calc GET #############################
py_calc_get = py_data['resource']['calc']['method']['GET'][0]
js_calc_get = js_data['resource']['calc']['method']['GET'][0]
##############################################################
####################### HTML GET #############################
py_html_get = py_data['resource']['html']['method']['GET'][0]
js_html_get = js_data['resource']['html']['method']['GET'][0]
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(15)
```

```python
def makePlot(ax,py_data,js_data,title):
    ax.plot(py_data, color = 'red')
    ax.plot(js_data, color = 'blue')
    
    ax.set_xlabel('Users',fontsize = 15)
    ax.set_ylabel('Time (ms)', fontsize = 15)
    ax.set_title(title, fontsize = 20)
```

```python
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
makePlot(ax1,py_json_get,js_json_get,'GET JSON')
makePlot(ax2,py_json_post,js_json_post,'POST JSON')
makePlot(ax3,py_calc_get,js_calc_get,'GET Calc')
makePlot(ax4,py_html_get,js_html_get,'GET html')
```

```python
fig.legend(['Pyton 3', 'JS (node)'])
fig.tight_layout()
fig
```
