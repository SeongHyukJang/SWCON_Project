```python
#https://www.youtube.com/watch?v=LosIGgon_KM
'''
querystring 요소들
v : video id
t : start time

querystring 만드는법

1. 변수 정의
ex.) qs = "v=" + "LoslGgon_KM"

2. urllib의 parse모듈 사용
'''
from urllib import parse
```

```python
dir(parse)
```

```python
params = {'v': "LoslGgon_KM"}
querystring = parse.urlencode(params)
querystring
```

```python
url = "https://www.youtube.com/watch" + "?" + querystring
resp = request.urlopen(url)
```

```python
resp.isclosed()
# 연결확인가능
```

```python
resp.code
```

```python
html = resp.read().decode('UTF-8')
```
