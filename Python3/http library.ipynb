```python
import lauda

timer = lauda.StopWatch()

timer.start()
x = http.client.HTTPConnection('www.khu.ac.kr')
x.connect()
timer.stop()

print(timer.elapsed_time)


timer.start()
x = http.client.HTTPConnection('www.khu.ac.kr')
x.connect()
x.request('GET', 'www.khu.ac.kr')
timer.stop()

print(timer.elapsed_time)
```
